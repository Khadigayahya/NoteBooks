```python

import os
import requests
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
```

```python

processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-large")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-large").to("cuda")
```

```python
!git clone https://github.com/IvonaTau/ikea.git
```

```python
dataset_path = 'ikea/images/objects'
```

```python
results = []
```

```python
# معالجة كل صورة في المجلد
for filename in os.listdir(dataset_path):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # يمكنك إضافة أنواع ملفات أخرى إذا لزم الأمر
        img_path = os.path.join(dataset_path, filename)
        raw_image = Image.open(img_path).convert('RGB')


        # معالجة الصورة بدون نص شرطي
        inputs = processor(raw_image, return_tensors="pt").to("cuda")
        out = model.generate(**inputs)
        caption_unconditional = processor.decode(out[0], skip_special_tokens=True)

        # تخزين النتائج
        results.append({
            'filename': filename,

            'caption_unconditional': caption_unconditional
        })

# طباعة النتائج
for result in results:
    print(f"Filename: {result['filename']}")
   #print(f"Conditional Caption: {result['caption_conditional']}")
    print(f"Unconditional Caption: {result['caption_unconditional']}")
    print('-' * 50)
```

```python
import pandas as pd

# تحويل النتائج إلى DataFrame
df = pd.DataFrame(results)

# حفظ النتائج في ملف CSV
df.to_csv('images_objects.csv', index=False)

print("تم حفظ النتائج في ملف captions_results.csv")
```
